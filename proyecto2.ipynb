```python
cat_df.nunique()
```

```python
def group_emails(df):
    # Mapping regional variants to a parent brand
    df['email_bin'] = df['P_emaildomain'].replace({
        'gmail.com': 'google', 'gmail': 'google',
        'hotmail.com': 'microsoft', 'outlook.com': 'microsoft', 'msn.com': 'microsoft', 'live.com.mx': 'microsoft',
        'yahoo.com': 'yahoo', 'yahoo.com.mx': 'yahoo', 'yahoo.fr': 'yahoo', 'ymail.com': 'yahoo',
        'anonymous.com': 'anonymous', 
        'protonmail.com': 'encrypted' # Often high risk in fraud
    })
    
    # Label everything else as "other"
    main_providers = ['google', 'microsoft', 'yahoo', 'anonymous', 'encrypted']
    df.loc[~df['email_bin'].isin(main_providers), 'email_bin'] = 'other'
    return df

data = group_emails(data)

data = pd.get_dummies(data, columns=['email_bin'], prefix='email')
```

```python
data = pd.get_dummies(data, columns=['ProductCD'], prefix='prod')
data = pd.get_dummies(data, columns=['card4'], prefix='card4_')
data = pd.get_dummies(data, columns=['card6'], prefix='card6_')
```

```python
num_vars = data.select_dtypes(include=['int64', 'float64']).copy()

num_vars.drop(columns=['TransactionID', 'isFraud'], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_vars_scaled = scaler.fit_transform(num_vars)

num_vars_scaled_df = pd.DataFrame(
    num_vars_scaled, 
    columns=num_vars.columns, 
    index=num_vars.index  # <--- THIS IS THE KEY
)
num_vars_scaled_df.head()
```

```python
num_vars_scaled_df.shape
```

```python
data[num_vars.columns] = num_vars_scaled
data_for_training = data.drop(["TransactionID", "P_emaildomain"], axis=1).copy()
```

# Clustering

```python
from sklearn.cluster import HDBSCAN
from sklearn.decomposition import PCA

y = data_for_training["isFraud"]
X = data_for_training.drop(["isFraud"], axis=1)

pca = PCA(n_components=0.90, random_state=42)
X_pca = pca.fit_transform(X)

hdb_pca = HDBSCAN(min_cluster_size=500, n_jobs=-1)
labels_pca = hdb_pca.fit_predict(X_pca)

pca_summary = pd.Series(labels_pca).value_counts().sort_index()

print(pca_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# (Assuming X_pca, labels_pca, and y are already defined from your code above)

# Create a figure with 2 subplots side-by-side
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# --- PLOT 1: HDBSCAN Clusters (Left) ---
# We use X_pca[:, 0] (1st component) and X_pca[:, 1] (2nd component) for the 2D plot
sns.scatterplot(
    x=X_pca[:, 0], 
    y=X_pca[:, 1], 
    hue=labels_pca, 
    palette='Spectral', # Good palette for multiple clusters
    alpha=0.6, 
    s=15,               # Size of the dots
    ax=ax[0]
)
ax[0].set_title('HDBSCAN Clusters\n(-1 represents Noise/Outliers)')
ax[0].set_xlabel('Principal Component 1')
ax[0].set_ylabel('Principal Component 2')
ax[0].legend(title='Cluster ID', bbox_to_anchor=(1.05, 1), loc='upper left')

# --- PLOT 2: Actual Fraud Labels (Right) ---
sns.scatterplot(
    x=X_pca[:, 0], 
    y=X_pca[:, 1], 
    hue=y, 
    palette={0: 'lightgrey', 1: 'red'}, # Make normal grey, and fraud stand out in red
    alpha=0.6, 
    s=15, 
    ax=ax[1]
)
ax[1].set_title('Actual Labels (Ground Truth)\n(0 = Normal, 1 = Fraud)')
ax[1].set_xlabel('Principal Component 1')
ax[1].set_ylabel('Principal Component 2')
ax[1].legend(title='isFraud', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
crosstab_df = pd.crosstab(
    index=labels_pca, 
    columns=y, 
    rownames=['HDBSCAN Cluster (-1 = Noise)'], 
    colnames=['Actual isFraud']
)

# Rename the columns for clarity (assuming 0 is Normal, 1 is Fraud)
crosstab_df.rename(columns={0: 'Normal (0)', 1: 'Fraud (1)'}, inplace=True)

# Add a Total column
crosstab_df['Total_Transactions'] = crosstab_df['Normal (0)'] + crosstab_df['Fraud (1)']

# Calculate the exact percentage of fraud in each cluster
crosstab_df['Fraud_Rate (%)'] = (crosstab_df['Fraud (1)'] / crosstab_df['Total_Transactions']) * 100

# Print the table nicely rounded
print(crosstab_df.round(2))
```

```python
data_for_training['hdbscan_cluster'] = labels_pca
```

# Reducción de Dimensionalidad

```python
import umap

reducer = umap.UMAP(
    n_neighbors=15, 
    min_dist=0.1, 
    n_components=2, 
    random_state=42,
    n_jobs=-1 
)

umap_embedding = reducer.fit_transform(X)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

sns.scatterplot(
    x=umap_embedding[:, 0], 
    y=umap_embedding[:, 1], 
    hue=data_for_training["hdbscan_cluster"], 
    palette='Spectral', 
    alpha=0.1,  
    s=2,        
    ax=ax[0]
)
ax[0].set_title('Full Dataset UMAP (CPU)\nColored by HDBSCAN Clusters')
ax[0].legend(bbox_to_anchor=(1.05, 1), loc='upper left', markerscale=5)

sns.scatterplot(
    x=umap_embedding[:, 0], 
    y=umap_embedding[:, 1], 
    hue=y, 
    palette={0: 'lightgrey', 1: 'red'}, 
    alpha=0.3, 
    s=2, 
    ax=ax[1]
)
ax[1].set_title('Full Dataset UMAP (CPU)\nColored by Actual Fraud')
ax[1].legend(title='isFraud', loc='upper right', markerscale=5)

plt.tight_layout()
plt.show()
```
